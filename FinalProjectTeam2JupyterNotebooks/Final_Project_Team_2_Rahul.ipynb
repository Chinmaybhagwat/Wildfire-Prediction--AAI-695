```python
import numpy as np
import pandas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

```python
data = pandas.read_csv('./data/training/test.csv')
data['eto'].fillna((data['eto'].mean()), inplace=True)
data['precipitation'].fillna((data['precipitation'].mean()), inplace=True)
data['solar_rad'].fillna((data['solar_rad'].mean()), inplace=True)
data['aver_vapor_press'].fillna((data['aver_vapor_press'].mean()), inplace=True)
data['max_air_temp'].fillna((data['max_air_temp'].mean()), inplace=True)
data['min_air_temp'].fillna((data['min_air_temp'].mean()), inplace=True)
data['aver_air_temp'].fillna((data['aver_air_temp'].mean()), inplace=True)
data['max_humidity'].fillna((data['max_humidity'].mean()), inplace=True)
data['min_humidity'].fillna((data['min_humidity'].mean()), inplace=True)
data['aver_humidity'].fillna((data['aver_humidity'].mean()), inplace=True)
data['dew_point'].fillna((data['dew_point'].mean()), inplace=True)
data['aver_wind_speed'].fillna((data['aver_wind_speed'].mean()), inplace=True)
data['wind_run'].fillna((data['wind_run'].mean()), inplace=True)
data['soil_temp'].fillna((data['soil_temp'].mean()), inplace=True)

no_fire = data.loc[data['has_fire']==0]
has_fire = data.loc[data['has_fire']==1]
class_count_0, class_count_1 = data['has_fire'].value_counts()
print(class_count_1, class_count_0)

has_fire_over = has_fire.sample(class_count_0, replace=True)
balanced_data = shuffle(pandas.concat([has_fire_over, no_fire]))
# no_fire_over = no_fire.sample(class_count_1, replace=True)
# balanced_data = shuffle(pandas.concat([no_fire_over, has_fire]))
train, test = train_test_split(balanced_data, test_size=0.2, random_state=32)

logisticRegr = LogisticRegression(max_iter=1000)
logisticRegr.fit(train.iloc[:, 2:-1], train.iloc[:, -1])
predictions = logisticRegr.predict(test.iloc[:, 2:-1])
cm = classification_report(test.iloc[:, -1], predictions)
print(cm)

tree = DecisionTreeClassifier(max_depth=10)
tree.fit(train.iloc[:, 2:-1], train.iloc[:, -1])
predictions2 = tree.predict(test.iloc[:, 2:-1])
cm2 = classification_report(test.iloc[:, -1], predictions2)
print(cm2)
```

```python
from sklearn.model_selection import GridSearchCV
parameters_logistic = {
    'penalty' : ['l2'], 
    'solver'  : ['newton-cg', 'lbfgs', 'liblinear'],
    'max_iter': [500, 1000]
}
logistic = LogisticRegression()
grid_search_logistic = GridSearchCV(logistic, param_grid = parameters_logistic, scoring='accuracy', cv=10, n_jobs=3)
grid_search_logistic.fit(train.iloc[:, 2:-1], train.iloc[:, -1])
logistic_best_param = grid_search_logistic.best_params_
print(logistic_best_param)
model_best_logistic=LogisticRegression(penalty=logistic_best_param['penalty'],
                                            solver=logistic_best_param['solver'],
                                            max_iter=logistic_best_param['max_iter'])
model_best_logistic.fit(train.iloc[:, 2:-1], train.iloc[:, -1])
pred_best_logistic = model_best_logistic.predict(test.iloc[:, 2:-1])
cm_best_log = classification_report(test.iloc[:, -1], pred_best_logistic)
print(cm_best_log)
```

```python
param_dt = {'criterion': ['gini', 'entropy'],
             'max_depth': range(5,10),
             'min_samples_split': range(2,10)}
dt = DecisionTreeClassifier()
grid_search_dt = GridSearchCV(dt, param_grid = param_dt, scoring='accuracy', cv=10, n_jobs=3)
grid_search_dt.fit(train.iloc[:, 2:-1], train.iloc[:, -1])
dt_best_param = grid_search_dt.best_params_
model_best_dt=DecisionTreeClassifier(criterion=dt_best_param['criterion'],
                                    max_depth=dt_best_param['max_depth'],
                                    min_samples_split=dt_best_param['min_samples_split'])
model_best_dt.fit(train.iloc[:, 2:-1], train.iloc[:, -1])
pred_best_dt = model_best_dt.predict(test.iloc[:, 2:-1])
cm_best_dt = classification_report(test.iloc[:, -1], pred_best_dt)
print(cm_best_dt)
```
